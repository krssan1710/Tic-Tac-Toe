```python
from IPython.display import clear_output

def display_index_positions_to_start_with():
    '''this is to print a matrix to familiarize users with index positions'''
    clear_output()
    print("Here we go!")
    print('These are the numbers corresponding to each position on the board')
    print('7'+'|'+'8'+'|'+'9')
    print('4'+'|'+'5'+'|'+'6')
    print('1'+'|'+'2'+'|'+'3')


def display(matrix):
    '''this is to display the matrix that is passed as an argument'''
    print("Current board status")
    print(matrix[2][0]+"|"+matrix[2][1]+"|"+matrix[2][2])
    print(matrix[1][0]+"|"+matrix[1][1]+"|"+matrix[1][2])
    print(matrix[0][0]+"|"+matrix[0][1]+"|"+matrix[0][2])

def get_player_marks():
    '''This is to let Player-1 choose his/her mark (X|O) and to assign the other mark to Player-2'''
    
    invalid_input = True
    while invalid_input:
        player1_mark = input("Player-1, please choose your mark (X or O): ").upper()
        #If input is not X or O loop back and ask for input again. If it is correct return value
        if player1_mark in ['X','x','O','o']:
            invalid_input = False
        else:
            print("Sorry, you can only choose X or 0: ")
    
    #print confirmation
    print("Player-1 has choosen {}".format(player1_mark))
    
    #If Player-1 chose X, give Player-2 O, or vice versa
    if player1_mark in ['X','x']:
        player2_mark = 'O'
    else:
        player2_mark = 'X'
    
    #print confirmation
    print('Player-2, you play with {}'.format(player2_mark))
    
    #return as a dictionary
    return {'Player1':player1_mark,'Player2':player2_mark}
    
def get_index_choice():
    '''This is to get the position of player choice'''
    while True:
        choice = input("Enter your position number (1-9): ")
        #If not digit loop back and ask for input
        if not choice.isdigit():
            #display(matrix)
            print("Enter only digits please")
        #If digit proceed to next check
        else:
            while True:
                #If input not in range, look back and ask for input
                if int(choice) not in range(1,10):
                    #display(matrix)
                    print("Enter only (1-9) please")
                    break
                #If input is digit and is in range, accept it and return corresponding matrix indices
                else:
                    if choice == '1':
                        return [0,0]
                    elif choice == '2':
                        return [0,1]
                    elif choice == '3':
                        return [0,2]
                    elif choice == '4':
                        return [1,0]
                    elif choice == '5':
                        return [1,1]
                    elif choice == '6':
                        return [1,2]
                    elif choice == '7':
                        return [2,0]
                    elif choice == '8':
                        return [2,1]
                    elif choice == '9':
                        return [2,2]

def is_index_empty(matrix,indices):
    '''This is to check is the index that a player chose is already marked'''
    if matrix[indices[0]][indices[1]] in [' ']:
        return True
    else:
        print("That position is already marked! Please choose an unmarked index")
        return False
                   
def mark_symbol_player(matrix,indices,player_mark):
    '''This is to mark the specific matrix index that user chooses'''
    matrix[indices[0]][indices[1]] = player_mark
    return(matrix)

def it_is_a_tie(matrix):
    '''This is to check if the game ended in a tie'''
    if ' ' not in matrix[0] and ' ' not in matrix[1] and not ' ' in matrix[2]:
        clear_output()
        print("It's a tie!")
        display(matrix)
        return True

def we_have_a_winner(matrix,mark):
    '''This is to check if there is a winner'''
    
    if ((matrix[0][0] == matrix[0][1] == matrix[0][2] == mark) or (matrix[1][0] == matrix[1][1] == matrix[1][2] == mark) or 
       (matrix[2][0] == matrix[2][1] == matrix[2][2] == mark) or (matrix[0][0] == matrix[1][0] == matrix[2][0] == mark) or 
       (matrix[0][1] == matrix[1][1] == matrix[2][1] == mark) or (matrix[0][2] == matrix[1][2] == matrix[2][2] == mark) or 
       (matrix[0][0] == matrix[1][1] == matrix[2][2] == mark) or (matrix[0][2] == matrix[1][1] == matrix [2][0] == mark)): 
            clear_output()
            print('******************We have a winner!******************')
            display(matrix)
            return True
    else:
        return False
    
def check_for_repeating_game():
    '''This is to check if user want to continue playing'''
    while True:
        choice = input("Do you want to play again? (Y or N): ")
        #if input not valid loop back and ask to enter again
        if choice not in ['Y','N','y','n'] :
            print("I dont understand, please enter (Y or N)")
        else:
            #return True or False based on input
            if choice == 'Y' or choice == 'y':
                return True
            else:
                return False

def play_game():
        '''This is to play the game'''
        #to hold indices corresponding to user position input
        indices = [0,0]
        
        #the game board
        matrix = [[' ',' ',' '],[' ',' ',' '],[' ',' ',' ']]
        
        #flag to keep track if there's a winner
        winner = False
        
        #flag to keep track if the game's a tie
        tie = False
        
        display_index_positions_to_start_with()
        player_marks = get_player_marks()
        
        #keep playing until there is a winner or until the game is a tie
        while not tie and not winner:
            
            #flag to check if a position ta marked already
            index_marked_already = False
            
            #Player-1's move
            #keep asking for marking position till an unmarked position is selected
            while not index_marked_already:
                print("Player-1's move")
                indices = get_index_choice()
                index_marked_already = is_index_empty(matrix,indices)
                
            #mark Player-1's symbol to the selected position
            matrix = mark_symbol_player(matrix,indices,player_marks['Player1'])
            clear_output()
            display(matrix)
            
            #check if Player-1's move resulted in a win or a tie
            winner = we_have_a_winner(matrix,player_marks['Player1'])
            tie = it_is_a_tie(matrix)
            
            #reset mark flag to re-use for Player-2
            index_marked_already = False
            
            #Player-2's move
            #get Player-2's input if Player-1's previous move did not end is a win or a tie
            if not tie and not winner:
                
                #keep asking for marking position till an unmarked position is selected
                while not index_marked_already:
                    print("Player-2's move")
                    indices = get_index_choice()
                    index_marked_already = is_index_empty(matrix,indices)
                
                #mark Player-2's symbol to the selected position
                matrix = mark_symbol_player(matrix,indices,player_marks['Player2'])
                clear_output()
                display(matrix)
                
                #check if Player-2's move resulted in a win or a tie
                winner = we_have_a_winner(matrix,player_marks['Player2'])
                tie = it_is_a_tie(matrix)
                
```

```python
game_on = True
#keep playing till players do not want to play anymore
while game_on:
    play_game()
    game_on = check_for_repeating_game()
clear_output()
#exit if players dont want to play anymore
print("Thanks for playing! Come back soon! :)")
```
